## Data Cleaning

Missing values will be dropped and filled
Reasons
- Most missing values have are very few compared to the total amount of rows like imdb_id which has 10 missing rows, cast has 76 missing rows, overview has 4 and genres has 3. These will be dropped

#### Movies produced in the 20th and 21st centuries and their increase over the years

```python
#filter the release years and group into 20th and 21st century
century_20 = df_mov[df_mov['release_year'] <= 2000]
century_21 = df_mov[df_mov['release_year'] > 2000]
```

```python
#using a line plot to visualize the amount of movies released in both centuries
century_20.groupby('release_year').count().original_title.plot(kind='line',figsize=(8,8))
plt.xlabel('Released Year')
plt.ylabel('No of Movies')
plt.title('Movies numbers trend in the 20th Century');
```

The line plot above shows an uneven increase and decrease in movie releases. There are sharp declines in years 1969 and 1990. There was little to no increase in movie release from 1990-1993, then a sharp increase in 1993. The uneven increase and decrease continues till 2000

```python
#using a line plot to visualize the amount of movies released in both centuries
century_21.groupby('release_year').count().original_title.plot(kind='line',figsize=(8,8))
plt.xlabel('Released Year')
plt.ylabel('No of Movies')
plt.title('Movies numbers trend in the 21st Century');
```

The line plot above shows a steady increase in movie releases from year 2001 but in 2009 there was a sharp decline but later picked up on a steady increase again in 2010. in 2014 the most movies were released compared to previous years, in 2015 a decline set in.



```python
#plot a line plot to show the movie releases in both centuries
df_mov.groupby('release_year').count().original_title.plot(kind='line', alpha=0.9,figsize=(20,8))
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.title('Movies numbers trend over two centuries');
```

#### Which genres are most popular from year to year?

```python
#use split() to split the genres since a row has multiple genres
#copy the dataframe into a new one
df_new = df_mov.copy()

#using split() function to seperate the genres in a single row
df_new['genres'] = df_new['genres'].str.split('|')

#a new row can be created to contain the splited genres using explode
df_new = df_new.explode('genres')

#filter to get only the popularity, genres and release year columns
pop_genre = df_new[['popularity', 'genres', 'release_year']]

#using groupby to get the most popular genres each year
pop_genre.groupby(['release_year', 'genres'])['popularity'].mean().groupby(level='release_year').nlargest(1).reset_index(level=0, drop=True)
```

```python
#create a line plot to view the movie release trend
df_new.groupby('release_year').popularity.mean().plot(kind='line',figsize=(8,8))
plt.title('Movie Popularity trend over the years')
plt.xlabel('Release Years')
plt.ylabel('Popularity Mean');
```

After using groupby to get the most popular genres each year, a line plot was used to display the release trend over the years which was not so consistent and steady. The popularity rate had its first sharp decline in year 1966 but came back up in 1967. This unsteady trend continued over the years till 2012 when the increase became constant.

### Do Movies with High budgets get Higher ratings

```python
df_mov.head()
```

```python
#creating a function to create a plot
def plot_analysis(data_frame, x_axis, y_axis, plt_kind, x_label, y_label, title):
    ''' Plots a graph'''
    data_frame.plot(x=x_axis, y=y_axis, kind=plt_kind, figsize=(12, 8)); 
    plt.xlabel(x_label); #x_axis title
    plt.ylabel(y_label); #y_axis title
    plt.title(title); #plot title
```

```python
#use the functioned created to plot a scatter plot
plot_analysis(df_mov, 'budget', 'vote_count', 'scatter', 'Budget', 'Ratings', 'Relationship Between Budget and Ratings')
```

The above plot indicates that budget does not determing the outcome of the rating. They are both independent of each other.
